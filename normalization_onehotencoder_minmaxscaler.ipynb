```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
threshold = 5

#Refining Data
categorical_cols_refined = [
    col for col in df.columns
    if len(df[col].unique()) <= threshold or df[col].dtype == 'object'
]
numerical_cols_refined = [col for col in df.columns if col not in categorical_cols_refined]

print("Refined Numerical Columns:", numerical_cols_refined)
print("Refined Categorical Columns:", categorical_cols_refined)

#Data imputation
for col in numerical_cols_refined:
    df[col] = df[col].fillna(df[col].mean())  # Impute with mean

for col in categorical_cols_refined:
    df[col] = df[col].fillna(df[col].mode()[0])  # Impute with mode
print(df.isnull().sum())
# Label encoding
for col in df.columns:
    if df[col].dtype == 'object':
        df[col]=le.fit_transform(df[col])



#df[['CAR','MaritalStatus']].sample(15)
```
